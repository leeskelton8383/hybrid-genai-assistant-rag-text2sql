Now that we have all the stats data downloaded from nflverse and cleaned, we'll load games, seasons, players into 3 tables in SQLlite

```python
import pandas as pd
import sqlite3
import os

# -----------------------------
# Config
# -----------------------------
csv_path = r"C:\Users\250331\OneDrive - MyFedEx\Documents\Python Scripts\EaglesGPT\data\stats_player_regpost_2020_2025.csv"
db_path = "data/stats.sqlite"
table_name = "players"

# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv(csv_path)

# Optional: check column names
print("Columns:", df.columns.tolist())
print("Number of rows:", len(df))

# -----------------------------
# Write to SQLite
# -----------------------------
os.makedirs("data", exist_ok=True)
conn = sqlite3.connect(db_path)

# This will drop and recreate the table if it exists
df.to_sql(table_name, conn, if_exists="replace", index=False)

conn.close()
print(f"âœ… Saved to {db_path} as table '{table_name}'")
```
